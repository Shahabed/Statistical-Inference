# Testing the `src` Package

This notebook demonstrates how to use the core reusable statistical modules in the `src/` package.

```python
import sys
sys.path.append('..')   # so we can import from src

import numpy as np
import pandas as pd

from src import bootstrap
from src import hypothesis_testing
from src import empirical_bayes
from src import shrinkage
from src import lasso
from src import survival
```

## 1. Bootstrap Module

```python
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)

def mean_stat(x):
    return np.mean(x)

# Bootstrap confidence interval
ci_low, ci_high = bootstrap.bootstrap_confidence_interval(
    data, mean_stat, n_bootstrap=5000, confidence_level=0.95
)
print(f"Bootstrap 95% CI for mean: [{ci_low:.2f}, {ci_high:.2f}]")

# Jackknife
jack_est, jack_se = bootstrap.jackknife_estimate(data, mean_stat)
print(f"Jackknife estimate: {jack_est:.2f} ± {jack_se:.2f}")
```

## 2. Hypothesis Testing Module

```python
np.random.seed(42)
sample1 = np.random.normal(100, 15, 50)
sample2 = np.random.normal(105, 15, 50)

t_stat, p_val = hypothesis_testing.two_sample_t_test(sample1, sample2)
print(f"Two-sample t-test: t = {t_stat:.3f}, p = {p_val:.4f}")

# Permutation test
obs_diff, p_perm = hypothesis_testing.permutation_test(sample1, sample2, n_permutations=2000)
print(f"Permutation test: observed diff = {obs_diff:.2f}, p = {p_perm:.4f}")
```

## 3. Empirical Bayes & Shrinkage

```python
np.random.seed(42)
true_means = np.array([10, 12, 8, 15, 9])
observations = true_means + np.random.normal(0, 3, 5)

shrunk_js = empirical_bayes.james_stein_estimator(observations)
print("James-Stein shrunk estimates:", np.round(shrunk_js, 2))

shrunk_positive = shrinkage.positive_part_james_stein(observations)
print("Positive-part James-Stein:", np.round(shrunk_positive, 2))
```

## 4. LASSO Module

```python
np.random.seed(42)
X = np.random.randn(100, 5)
true_beta = np.array([3, 0, -2, 0, 1.5])
y = X @ true_beta + np.random.normal(0, 1, 100)

coef, model = lasso.fit_lasso_sklearn(X, y, alpha=0.5)
print("LASSO coefficients:", np.round(coef, 3))
```

## 5. Survival Analysis Module

```python
np.random.seed(42)
durations = np.random.exponential(scale=10, size=50)
events = np.random.binomial(1, 0.7, 50)   # 1 = event occurred

km_table = survival.kaplan_meier_estimator(durations, events)
print(km_table.head())

median_surv = survival.kaplan_meier_median_survival(durations, events)
print(f"Median survival time: {median_surv:.2f}")
```
